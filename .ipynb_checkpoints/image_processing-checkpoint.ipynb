```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
"""
Image filtering using Non-local Means Denoising
"""

rawdata = cv2.imread('00025A545-PSTM_Flatbed.png')

nlmean_filtered_data = cv2.fastNlMeansDenoising(rawdata,None,50,50,7)
#dst = cv2.fastNlMeansDenoisingColored(img,None,50,50,7,21)

plt.figure(figsize=(25,10))
plt.subplot(121),plt.imshow(rawdata)
plt.subplot(122),plt.imshow(nlmean_filtered_data)
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

##img = cv2.imread('messi5.jpg',0)
edge_detect_data = cv2.Canny(nlmean_filtered_data,300,100)

plt.figure(figsize=(25,10))
plt.subplot(121),plt.imshow(nlmean_filtered_data,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edge_detect_data,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
edge_detect_data
```

```python
"""
Applying Gabor filter
"""
kernel = cv2.getGaborKernel((11, 11), 4, 11, 11, 1, 1, cv2.CV_32F)
#kernel /= math.sqrt((kernel * kernel).sum())
kernel /= np.sqrt((kernel * kernel).sum())
gabor_filtered_data = cv2.filter2D(rawdata, -1, kernel)

plt.figure(figsize=(25,10))
plt.subplot(131),plt.imshow(rawdata)
plt.title('Raw Image')
plt.subplot(132),plt.imshow(nlmean_filtered_data)
plt.title('nlmean_filtered_data Image')
plt.subplot(133),plt.imshow(gabor_filtered_data)
plt.title('Gabor Filtered')
plt.show()
```

```python
"""
Edge detection on top of Gabor filter
"""
edge_detect_gabordata = cv2.Canny(gabor_filtered_data,300,100)

plt.figure(figsize=(25,10))
plt.subplot(131),plt.imshow(rawdata)
plt.title('Raw Image')
plt.subplot(132),plt.imshow(edge_detect_data,cmap='gray')
plt.title('edge_detect_data Image')
plt.subplot(133),plt.imshow(edge_detect_gabordata,cmap='gray')
plt.title('edge_detect_gabordata')
plt.show()
```
